# map 

uma função é aplicada a cada elemento de uma lista resultando em uma nova lista de mesmo comprimento que a original.

```python
#lista orignal
frutas = ["laranja","abacaxi","pêssego","morango"]

# função que será usada para a operação de 'map'
def para_maiuscula(palavra):
    return palavra.upper()
```

aplicando a função a cada elemento da lista usando um loop `for` (para cada)

```python
for palavra in frutas:
    frutas_maiuscula = para_maiuscula(palavra)
    print(frutas_maiuscula)
```

mas isso não retorna uma lista!

```python
def frutas_em_maiusculas(lista_de_frutas):
    for fruta in lista_de_frutas:
        fruta_maiuscula = para_maiuscula(fruta)
        print(fruta_maiuscula)
```

```python
# essa função vai fazer um 'map' propriamente
# porque ele vai criar uma NOVA lista mapenado 
# os dados da lista de entrada
def frutas_maiusculas(lista_de_frutas):
    nova_lista = list()
    for fruta in lista_de_frutas:
        fruta_maiuscula = para_maiuscula(fruta)
        nova_lista.append(fruta_maiuscula)
    return nova_lista
```

```python
frutas_em_maiusculas(frutas)
```

## map()

```python
#lista orignal
frutas = ["laranja","abacaxi","pêssego","morango"]

# função que será usada para a operação de 'map'
def para_maiuscula(palavra):
    return palavra.upper()
```

```python
list(map(para_maiuscula, frutas))
```

```python
for fruta_maiusculas in (map(para_maiuscula, frutas)):
    print(fruta_maiusculas)
```

além disso, a função `frutas_em_maiusculas()` tem um efeito colateral : ela não apenas processa os dados da lista como tambem escreve na tela 

### Exemplo:

map com a função `f(x) = x + 2`

```python
valores  = [1,2,3,4,5]

def f(x):
    return x + 2
list(map(f, valores))
```

função dobro(x) = x*2

```python
def dobro(x):
    return x * 2
list(map(dobro, valores))
```
